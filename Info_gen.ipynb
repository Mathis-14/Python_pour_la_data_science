Exercice 1

Generate $X$ a random variable, 1000 repetitions of a $U(0,1)$ distribution * $Y$ a random variable, 1000 repetitions of a normal distribution with zero mean and variance equal to 2 * Verify the variance of $Y$ with np.var

```python
import numpy as np
import numpy.random as rd


X=rd.random(1000)
Y=rd.normal(0,np.sqrt(2),1000)
print(np.var(Y))
```

```python
import matplotlib.pyplot as plt
plt.hist(X)
plt.show()
```

```python
plt.hist(Y)
print(np.var(Y))
plt.show()
```
